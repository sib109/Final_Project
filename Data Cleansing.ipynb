```python
import pandas as pd
import os
import csv
```

```python
alcohol_input_df = pd.read_excel('Data/Input/Alcohol.xlsx')
alcohol_input_df = alcohol_input_df[['State','Location', '2005 Both Sexes','2006 Both Sexes',
                                     '2007 Both Sexes','2008 Both Sexes','2009 Both Sexes', 
                                     '2010 Both Sexes', '2011 Both Sexes']]
```

```python
alcohol_input_df.rename(columns = {'Location':'County', '2005 Both Sexes':'2005','2006 Both Sexes':'2006','2007 Both Sexes':'2007',
                                  '2008 Both Sexes':'2008','2009 Both Sexes':'2009','2010 Both Sexes':'2010',
                                  '2011 Both Sexes':'2011',}, inplace = True)
```

```python
years = list(alcohol_input_df.columns)
years = years[2:]
```

```python
alcohol_clean_df = []
for year in years:
    input_df = alcohol_input_df[['County', 'State', year]]
    input_df['Year'] = year
    input_df['Alcohol'] = input_df[year]
    alcohol_clean_df.append(input_df[['County', 'State', 'Year', 'Alcohol']])
alcohol_clean_df = pd.concat(alcohol_clean_df)
```

```python
alcohol_clean_df[alcohol_clean_df.County != 'United States']
alcohol_clean_df['County State'] = alcohol_clean_df['County']+' '+ alcohol_clean_df['State']
alcohol_clean_df.to_excel('Data/Output/AlcoholClean.xls')
```
