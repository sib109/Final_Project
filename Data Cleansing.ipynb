```python
obesity_input_df = pd.read_csv('Data/Input/Obesity.csv')
obesity_df = obesity_input_df[['merged_fips', 'State', 'County', 'Sex','Prevalence 2005 (%)', 'Prevalence 2006 (%)',
                  'Prevalence 2007 (%)', 'Prevalence 2008 (%)', 'Prevalence 2009 (%)', 'Prevalence 2010 (%)','Prevalence 2011 (%)']]
obesity_df_m = obesity_df.loc[obesity_df['Sex'] == 'Male',:] 
obesity_df_f = obesity_df.loc[obesity_df['Sex'] == 'Female',:] 
obesity_df_merged = pd.merge(obesity_df_m, obesity_df_f, on = 'merged_fips')
obesity_df_merged ['2005'] = (obesity_df_merged ['Prevalence 2005 (%)_x']+ obesity_df_merged ['Prevalence 2005 (%)_y'])/2
obesity_df_merged ['2006'] = (obesity_df_merged ['Prevalence 2006 (%)_x']+ obesity_df_merged ['Prevalence 2006 (%)_y'])/2
obesity_df_merged ['2007'] = (obesity_df_merged ['Prevalence 2007 (%)_x']+ obesity_df_merged ['Prevalence 2007 (%)_y'])/2
obesity_df_merged ['2008'] = (obesity_df_merged ['Prevalence 2008 (%)_x']+ obesity_df_merged ['Prevalence 2008 (%)_y'])/2
obesity_df_merged ['2009'] = (obesity_df_merged ['Prevalence 2009 (%)_x']+ obesity_df_merged ['Prevalence 2009 (%)_y'])/2
obesity_df_merged ['2010'] = (obesity_df_merged ['Prevalence 2010 (%)_x']+ obesity_df_merged ['Prevalence 2010 (%)_y'])/2
obesity_df_merged ['2011'] = (obesity_df_merged ['Prevalence 2011 (%)_x']+ obesity_df_merged ['Prevalence 2011 (%)_y'])/2
obesity_df2 = obesity_df_merged[['merged_fips', 'State_x', 'County_x','2005', '2006', '2007', '2008', '2009', '2010','2011']]
years = list(obesity_df2.columns)
years = years[3:]
obesity_df = []
for year in years:
    input_df = obesity_df2[['merged_fips', 'State_x', 'County_x', year]]
    input_df['Year'] = year
    input_df['Obesity'] = input_df[year]
    obesity_df.append(input_df[['merged_fips', 'State_x', 'County_x', 'Year', 'Obesity']])
obesity_df = pd.concat(obesity_df)
obesity_df['ModifiedFIPS'] = obesity_df['merged_fips'].apply(lambda x: '0'+str(x) if len(str(x)) < 5 else str(x))
obesity_df[['Year']] = obesity_df[['Year']].astype(str) 
obesity_df['FIPS Year Key'] = obesity_df['ModifiedFIPS']+"-"+obesity_df['Year']
obesity_df.head()
obesity_df.rename(columns = {'County_x':'County', 'State_x':'State'}, inplace = True)
obesity_final_df = obesity_df[['County', 'State', 'ModifiedFIPS','Year', 'FIPS Year Key','Obesity']]
obesity_final_df.to_excel('Data/Output/ObesityClean.xlsx', index = False)
```
